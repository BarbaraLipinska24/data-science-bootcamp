```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
fig=plt.figure()
fig.show()
```

1. Wykres liniowy

```python
plt.plot([0,1],[0,3])
```

```python
plt.plot([0,1],[0,3],label='line',color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_=plt.legend()
```

```python
plt.plot([0,1,2],[0,3,2],label='line',color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_=plt.legend()
```

```python
x = np.arange(-3,3,0.1)
x_sin=np.sin(x)
x_cos=np.cos(x)
plt.plot(x,x_sin)
_=plt.plot(x,x_cos)
```

2. Tworzenie wielu wykresów

```python
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,3,2],[3,2,5], 'r--')
```

```python
plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,3,2],[3,2,5], 'r--')
plt.subplot(133)
plt.plot([0,3,2],[3,2,5], 'ro')
```

3. Wykresy słupkowe

```python
plt.bar(x=[0,1,2,3], height=[1.2,3.2,1.2,5.2])
```

```python
plt.bar(x=['a','b','c','d'], height=[1.2,3.2,1.2,5.2])
```

```python
plt.barh(y=['a','b','c','d'], width=[1.2,3.2,1.2,5.2])
```
